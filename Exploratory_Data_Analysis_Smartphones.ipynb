## **Data Manipulation**

Before conducting exploratory data analysis (EDA), it's essential to perform data manipulations to update the data types of specific columns. This ensures seamless EDA operations and facilitates accurate analysis. We'll focus on updating the data types of some of the columns to ensure consistency and enhance our analytical capabilities.

As you can see, the price columns currently contain values stored as strings and include the rupee symbol, which prevents us from performing numerical analysis. To address this, we need to clean and convert these values into a numerical format, enabling us to use this column for further data analysis and insights.

#### **Chnaging Data Types**


```python
df.price
```

```python
df.price = df['price'].str.replace(',','').str.replace('₹','').astype(int)
df.price
```

#### **Checking and Handling missing values**

Within our rating column, several null values have been detected, posing potential obstacles for our analysis. To ensure data completeness and facilitate seamless analysis, we will replace these null values with the mean rating value.

```python
df.rating.isna().value_counts()
```

```python
median_rating = round(df.rating.mean(),2)
median_rating
```

```python
df.rating.fillna(median_rating,inplace = True)
```

```python
df.rating.isna().value_counts()
```

#### **Adding Extra columns for the Analysis**

We're planning to enrich our dataset by adding two condition-based columns.  
- The first column will categorize smartphone models into four segments based on their price, labeled as segments "Entry Level", "Economy Segment", "Business Segment", and "Premium Segment".  
- The second column, based on ratings, will classify smartphone models into three categories: high, medium, and low.

For this task, we've developed two functions, price_category() and rating_category(). These functions will facilitate the process of categorizing smartphone models based on price and ratings, respectively

```python
def price_catgeory(price) :
    if price < 20000 :
        return "Entry Level"
    elif price >= 20000 and price < 40000:
        return "Economy Segment"
    elif price >= 40000 and price < 60000 :
        return "Business Segment"
    else :
        return "Premium Segment"
```

```python
def rating_catgeory(rating) :
    if rating < 75 :
        return "Low"
    elif rating >= 75 and rating < 85:
        return "Medium"
    else :
        return "High"
```

```python
df["Price_Category"] = df['price'].apply(price_catgeory)
```

```python
df["Rating_Category"] = df['rating'].apply(rating_catgeory)
```

In today's world, nearly all smartphones boast fast charging capabilities. In our dataset, the 'battery' column provides information not only about the battery capacity but also indicates whether the smartphone supports fast charging.
- To enhance our analysis, we'll introduce an additional column to specifically identify whether each smartphone model supports fast charging or not.

```python
df['Suppports Fast Charging'] = df.battery.apply(lambda x: 'Yes' if 'Fast Charging' in x else 'No')
```

**RAM**, or Random Access Memory, is a super-fast type of memory in mobile devices. It stores the apps you've opened since you turned on your phone, allowing you to seamlessly return to previously used apps without delay.
- To leverage this crucial aspect of mobile performance, I utilized the 'ram' column to extract RAM values of smartphones. Subsequently, I created a new 'RAM_in_GB' column to represent this vital memory metric in our dataset. This enhancement allows for more comprehensive analysis and understanding of smartphone specifications.

```python
split_values = df.ram.str.split(',',expand = True)
df['RAM_in_GB'] = split_values[0]
```

- In the 'sim' column of our dataset, we found comprehensive information about various connectivity features of smartphones, including 5G, NFC, and IR blaster support.
- To streamline our analysis and make these features more accessible, we created three new columns: '5G Support', 'NFC Support', and 'IR Blaster Support'.
- These columns indicate whether each smartphone model supports these features or not.
- This enhancement enables us to quickly identify smartphones with specific connectivity capabilities and perform targeted analyses on these attributes.

```python
df['5G Support'] = df.sim.apply(lambda x: 'Yes' if '5G' in x else 'No')
df['IR Blaster Support'] = df.sim.apply(lambda x: 'Yes' if 'IR Blaster' in x else 'No')
df['NFC Support'] = df.sim.apply(lambda x: 'Yes' if 'NFC' in x else 'No')
```
